```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from skimage.io import imread, imread_collection
from skimage.filters import scharr

from tqdm import tqdm
%matplotlib inline
```

```python
TRAIN_PATH = 'input/train'
TEST_PATH = 'input/test'
FILE_PATH = 'output/'
TRAIN_IMAGE_PATTERN = "%s/{}/images/{}.png" % TRAIN_PATH
TRAIN_MASK_PATTERN = "%s/{}/masks/*.png" % TRAIN_PATH
TEST_IMAGE_PATTERN = "%s/{}/images/{}.png" % TEST_PATH
```

```python
def read_img(img_id, flag_train=True):
    if flag_train:
        img_path = TRAIN_IMAGE_PATTERN.format(img_id, img_id)
    else:
        img_path = TEST_IMAGE_PATTERN.format(img_id, img_id)
    img = imread(img_path)
    img = img[:, :, :3]
    return img
```

```python
def read_mask(mask_id):
    mask_path = TRAIN_MASK_PATTERN.format(mask_id, mask_id)
    masks = imread_collection(mask_path).concatenate()
    _,height, width = masks.shape
    num_masks = masks.shape[0]
    mask = np.zeros((height, width), np.uint32)
    c = 3
    for index in range(0, num_masks):
        dm = scharr(masks[index])
        mask[dm > 0] = c
        if c!=999:
            c += 3
        else:
            c = 3
    return mask
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
ids = train.loc[train.hsv_cluster==0, 'image_id'].values
```

```python
imgs = [read_img(i) for i in ids]
```

```python
masks = [read_mask(i) for i in tqdm(ids)]
```

```python
_,_ = plt.subplots(1,1,figsize=(10,10))
plt.imshow(masks[3])
```

```python
stack = []
stack0 = []

for i in tqdm(range(len(imgs))):
    img = imgs[i]
    mask = masks[i]
    for m in mask:
        cimg = img.copy().astype(float)
        m = np.dstack([m,m,m])
        cimg[m==0] = np.nan
        stack.append([np.nanmean(cimg[:,:,0])/3+np.nanmean(cimg[:,:,1])/3+np.nanmean(cimg[:,:,2])/3])
        cimg = img.copy().astype(float)
        cimg[m!=0] = np.nan
        stack0.append([np.nanmean(cimg[:,:,0])/3+np.nanmean(cimg[:,:,1])/3+np.nanmean(cimg[:,:,2])/3])
```

```python
stack = np.array(stack)
stack0 = np.array(stack0)
```

```python
sns.distplot(np.log(stack)-np.log(stack0), label='kernel')
plt.legend()
```

```python
ids = train.loc[train.hsv_cluster==3, 'image_id'].values
```

```python
imgs = [read_img(i) for i in ids]
```

```python
masks = [read_mask(i) for i in ids]
```

```python
rstack = []
gstack = []
bstack = []
stack1 = []

for i in tqdm(range(len(imgs))):
    img = imgs[i]
    mask = masks[i]
    for m in mask:
        cimg = img.copy().astype(float)/255
        m = np.dstack([m,m,m])
        cimg[m==0] = np.nan
        rstack.append(np.nanmean(cimg[:,:,0]))
        gstack.append(np.nanmean(cimg[:,:,1]))
        bstack.append(np.nanmean(cimg[:,:,2]))
        stack1.append(np.nanmean(cimg[:,:,0])/3+np.nanmean(cimg[:,:,1])/3+np.nanmean(cimg[:,:,2])/3)
```

```python
sns.distplot(rstack, label='r')
sns.distplot(gstack, label='g')
sns.distplot(bstack, label='b')
sns.distplot(stack1, label='all')
plt.legend();
```
